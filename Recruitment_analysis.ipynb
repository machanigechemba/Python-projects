```python
#copied code a little modification by self more
#function to format the strings into short form
def format_df(df, column, palette):
  str_list = []
  for i in df[column]:
    if pd.isnull(i):
            continue
    data = i.split(";")
    for j in data:
      str_list.append(j)
  newdf = pd.DataFrame(str_list, columns=[column])
  data = newdf[column].value_counts()
  keys = newdf[column].value_counts().keys()
  percent = (data/df[column].notnull().sum())*100
  txt = [str(round(i,1))+'%' for i in percent]
  
  plt.figure(figsize=(11, 7))
  colors = ['brown' if ((x < 3500) & (x > 3100))  else 'grey' for x in data] #to color bars 16 & 17 differently
  ax = sns.barplot(y = keys, x = data, orient = 'h', palette = colors)
  #ax.bar_label(container = ax.containers[0], labels=data, label_type = "edge", color='black', fontsize=10) #label each bar with % of the total
   # Add percentage labels to the bars
  for i, (value, label) in enumerate(zip(data.values, data)):
        ax.text(value, i, f' {label}', va='center')
      
  for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False) #remove all grids
  #move the tick labels to the top of the graph using the Axes.xaxis.tick_top() method:
  # remove the top ticks instead of the bottom ones
  # color the x-tick labels, we use the same Axes.tick_params()
  ax.xaxis.tick_top()
  ax.xaxis.set_label_position('top')
  ax.tick_params(top=False, left=False)
  plt.tight_layout()

  return ax
```

```python
#function to format strings. split by space into words and take the first 2 words
def string_format(df, column, color):
    str_list = []
    for i in df[column]:
        if pd.isnull(i):
            continue
        # Split the entry by space and take the first two words
        words = i.split()[:2] #split the string at space into words and take the first two words
        data = " ".join(words) #join the two words together again to create a shorter string
        str_list.append(data) #append the shorter string to the empty list

    # Create a new DataFrame from the list of first two words
    newdf = pd.DataFrame(str_list, columns=[column])

    # Count occurrences of each unique value
    data_counts = newdf[column].value_counts()

    # Calculate percentages
    #percent = (data_counts / df[column].notnull().sum()) * 100

    # Create a list of percentage labels. #get value_counts as percentages by setting normalize=True as str
    text = df[column].value_counts(normalize=True).mul(100).round(2).astype(str)+'%'

    # Create the plot
    sns.set_theme()
    plt.figure(figsize=(8, 4))
    ax = sns.barplot(x=data_counts.values, y=data_counts.index, hue=data_counts.index, dodge=False, palette=color)

    # Add percentage labels to the bars
    for i, (value, label) in enumerate(zip(data_counts.values, text)):
        ax.text(value, i, f' {label}', va='center')

    # Set plot labels and title
   # ax.set_xlabel('Counts')
    #ax.set_ylabel(column)
   # ax.set_title(f'Bar plot of {column}')
    ax.legend().set_visible(False)
    [spine.set_visible(False) for spine in ax.spines.values()] #remove all spines
    ax.xaxis.tick_top()
    ax.grid(False)
    ax.set_facecolor('white')
    ax.tick_params(left=False, top = False)
    plt.tight_layout() #this makes sure that all objects are inside the fig boundaries
    return ax
```

```python
#testing
fig1 = string_format(data, "EdLevel", "rocket")
fig1.set_xticks([0, 250, 500, 750, 1000, 1250, 1500, 1700, 2000])
fig1.set_xticklabels(['0', '250', '500', '750', '1,000', '1,250', '1,500', '1,700', '2,000'], color = 'black')
fig1.set_xlabel(' ')
fig1.set_ylabel(' ')
fig1.set_title("Data Scientists have Bachelor's or Master's Degree", y = 1.17, x = 0.45, weight = "bold", fontsize = 12)
plt.suptitle("Most data scientists/analysts have a degree but only 10.6% hold a PhD", y = 1.02, x = 0.58, fontsize = 10)
plt.savefig('education level.png')
plt.show()
```

#### Format the currency and compensation frequency columns

```python
# Function to split the currency info, extract the currency code for not null values only
def split_currency_info(df, col):
    df.loc[:, col] = df[col].str.split('\t').str[0]
    for string in df[col]:
      if pd.isnull(string):
        continue
    return df[col].head(5)
```

```python
#let's find the most frequent currencies for each country# Group by 'country' and get the most common 'currency'

country_currency_mode = data.groupby('Country')['Currency'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None).reset_index()
```

```python
#Merge the grouped data with the original DataFrame to bring in the most common currency for each country.

data = pd.merge(data, country_currency_mode, on='Country', suffixes=('', '_mode'))
```

```python
# Convert the compensation Frequency to yearly
def convertcomp_freq(x):
    x = str(x)
    if x == 'Monthly':
        return 12
    elif x == 'Weekly':
        return 52
    elif x == 'Yearly':
        return 1
    else:
        return 1 #assume where there is no value is yearly


data.loc[:, 'CompFreq'] = data['CompFreq'].apply(lambda x: convertcomp_freq(x))
```

salary_in_usd = CompTotal, which I have already converted to annual salary divide by the exchange rate.

The value column shows how much of the other currency is equal to one dollar (base currency is USD). e.g., 1 USD = 4.1 ILS / 0.8 GBP

```python
# Calculate salary in USD
merged_df['salary_in_usd'] = merged_df['CompTotal'] / merged_df['Value']
```

```python
merged_df[['Currency', 'CompTotal', 'Value', 'CompFreq', 'salary_in_usd']].head(10)
```

```python
# Check for conversion errors
merged_df['conversion_error'] = merged_df['salary_in_usd'] != merged_df['salary_in_usd']
# Display rows with conversion errors
conversion_errors = merged_df[merged_df['conversion_error']]
print(conversion_errors)
```

```python
# Re-check the minimum salary rows
min_salary_values = merged_df['salary_in_usd'].nsmallest(2)
min_salary_rows = merged_df[merged_df['salary_in_usd'].isin(min_salary_values)]

print("Rows with minimum salary values:")
print(min_salary_rows)
```

```python
low_salary = merged_df.query('salary_in_usd < 1000')
low_salary[['Country', 'Currency', 'CompTotal', 'CompFreq', 'Value', 'salary_in_usd']]
```

Some of the lowest salary is in GBP and USD in the original columns. These are more likely in thousands.
Convert them to thousands

```python
# Identify the rows with small values in salary_in_usd and original currency is USD or GBP

mask = (merged_df['salary_in_usd'] < 1000) & (merged_df['Currency'].isin(['USD', 'GBP']))

# Update the salary_in_usd values by multiplying by 1000

merged_df.loc[mask, 'salary_in_usd'] *= 1000
```

Removes rows where the annual compensation is less than 1k or greater than 500k USD, as these values might be unrealistically low/high or indicative of incorrect data entry.

```python
# remove outliers
outliers_small = merged_df.salary_in_usd < 1000
outliers_big = merged_df.salary_in_usd > 500000
merged_df = merged_df[~outliers_small & ~outliers_big].copy() #selects only the rows from df where salary_in_usd is not in the outliers list.
```

```python
#change dtype from float64 to float32 to solve memory error
# Select columns with 'float64' dtype  

float64_cols = list(merged_df.select_dtypes(include='float64'))

# The same code again calling the columns

merged_df[float64_cols] = merged_df[float64_cols].astype('float32')
```
